```python
import pandas as pd

# Function to load a CSV file with error handling
def load_data(url):
    try:
        data = pd.read_csv(url)
        print("Data loaded successfully.")
        return data
    except Exception as e:
        print(f"Error loading data from {url}: {e}")
        return None

# Load the drinks dataset
drinks_url = 'http://bit.ly/drinksbycountry'
drinks = load_data(drinks_url)

if drinks is not None:
    # Display the shape of the DataFrame
    print("Shape: ", drinks.shape)
    print()

    # Current max_rows option
    print("Current max_rows option:", pd.get_option('display.max_rows'))
    print(drinks)
    print()

    # Set max_rows to 2 for display
    pd.set_option('display.max_rows', 2)
    print(drinks)
    print()

    # Reset max_rows to default
    pd.reset_option('display.max_rows')
    print(drinks)
    print()

    # Create new columns based on existing data
    drinks['x'] = drinks.wine_servings * 1000
    drinks['y'] = drinks.total_litres_of_pure_alcohol * 1000

    # Display the first few rows of the modified DataFrame
    print(drinks.head())
    print()

    # Set float format for display
    pd.set_option('display.float_format', '{:,.2f}'.format)  
    print(drinks.head())
    print()

    # Load the training dataset
    train_url = 'http://bit.ly/kaggletrain'
    train = load_data(train_url)

    if train is not None:
        # Current max_colwidth option
        print("Current max_colwidth option:", pd.get_option('display.max_colwidth'))
        print(train.head())
        print()

        # Set max_colwidth to 1000 for display
        pd.set_option('display.max_colwidth', 1000)
        print(train.head())
        print()
```
